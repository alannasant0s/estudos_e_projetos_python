Exercício 1.1: Crie uma função que recebe uma lista de números e retorne:

O maior número

O menor número

A média aritmética


```python
def analise_numeros(lista):
    # Implemente aqui
    maior = max(lista)
    menor = min(lista)
    media = sum(lista) / len(lista)
    return maior, menor, media, f"O maior número é {maior}, o menor número é {menor}, e a média é {media:.1f}"
# Teste
print(analise_numeros([5, 2, 9, 1, 7]))  # Deve retornar (9, 1, 4.8)
```

```python
def analise_numeros(lista):
    maior = max(lista)
    menor = min(lista)
    media = sum(lista) / len(lista)

    # Retornando tanto os valores quanto uma string formatada
    return maior, menor, media, f"O maior número é {maior}, o menor número é {menor}, e a média é {media:.1f}"

# Teste
resultado = analise_numeros([5, 2, 9, 1, 7])
print(resultado[3])  # Imprime a mensagem formatada
```
